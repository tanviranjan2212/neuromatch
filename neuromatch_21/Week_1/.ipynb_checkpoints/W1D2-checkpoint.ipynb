#### Section 1.1: Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch import nn
from math import pi
import seaborn as sns
```

```python
def fun_z(x,y):
    return np.sin(x**2+y**2);
def fun_dz(x,y):
    dz_dx = np.cos(x**2 + y**2)*2*x;
    dz_dy = np.cos(x**2 + y**2)*2*y;
    return dz_dx, dz_dy
    
x, y = np.arange(-3, 3.01, 0.02), np.arange(-3, 3.01, 0.02)
xx, yy = np.meshgrid(x, y, sparse=True)
zz = fun_z(xx, yy)
plt.contourf(x,y,zz); 

xg, yg = np.arange(-2.5, 2.6, 0.5), np.arange(-2.5, 2.6, 0.5)
xxg, yyg = np.meshgrid(xg, yg, sparse=True)
zxg, zyg = fun_dz(xxg, yyg)
plt.quiver(xg, yg, zxg, zyg, scale=50, color ='r')
plt.title('Gradient for $z=sin(x^2 + y^2)$');

plt.show();
```

Set the seed to maintain reproducibility

```python
import random
import torch

def set_seed(seed=None, seed_torch=True):
    if seed is None:
        seed = np.random.choice(2 ** 32)
    random.seed(seed)
    np.random.seed(seed)
    if seed_torch:
        torch.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.cuda.manual_seed(seed)
        torch.backends.cudnn.benchmark = False
        torch.backends.cudnn.deterministic = True

    print(f'Random seed {seed} has been set.')


# In case that `DataLoader` is used
def seed_worker(worker_id):
    worker_seed = torch.initial_seed() % 2**32
    np.random.seed(worker_seed)
    random.seed(worker_seed)
```

#### Section 1.2: Pytorch Autograd
Automatic differentiation is honestly pretty great! The software builds computational graphs without much info needed from the user.

In Pytorch, we can set requires_grad=True to tell the software that the tensor contains learnable parameters. Pytorch will then keep track of every computation (differentiation) done with this tensor while making the computational graph.

Let's build a simple computational graph for the following function
$$ y = tanh(wx+b)$$,
$$ Loss = (y - tanh(wx+b))**2$$

```python
class Graph_struct:
    def __init__(self, w, b):
        self.w = torch.tensor([w],requires_grad=True);
        self.b = torch.tensor([b],requires_grad=True);
    def forward(self, x):
        return torch.tanh(self.w*x+self.b);
def sq_loss(y_true, y_pred):
    return (y_true - y_pred)**2

x=torch.tensor([7]); 
y_true=torch.tensor([1]);

graph1 = Graph_struct(-0.5, 0.5)
y_pred = graph1.forward(x);
loss = sq_loss(y_true, y_pred);
print(f'x:{x.item()}, $y_{{true}}$: {y_true.item()}, $y_{{pred}}$: {y_pred.item():0.2f}, L2 loss = {loss.item():0.2f}')

```

Some useful info about automatic differentiation:
* **detach() function**:
Learnable parameters (those with requires_grad=True) are contagious, ie., if $y=w*x$, and w is learnable, then y is also learnable. So any operation that is applied to y will also be in the computational graph. If we needed to extract the computational graph related to y, we must detach it by using the .detach on the variable, as shown in the following example

* **backward()**:
This accumulates gradients in the leaf nodes

* **dir(graph1.w)**:
We can use dir(graph_variable) to show the computational graph for each variable

Let's calculate derivatives by hand, before we use the .backward() function

```python
# Analytical derivative
dloss_dw = -2 * x * (y_true - y_pred.detach())*(1 - y_pred.detach()**2)
dloss_db = -2 * (y_true - y_pred.detach())*(1 - y_pred.detach()**2)

# Autograd derivative
loss.backward(retain_graph=True)
autograd_dloss_dw = graph1.w.grad
autograd_dloss_db = graph1.b.grad

print('Analytical derivative vs autograd derivative - they match!');
print(f'{dloss_dw} vs {autograd_dloss_dw}')
print(f'{dloss_db} vs {autograd_dloss_db}')
```

##### Section 1.3: Pytorch's nn module
Next, let's generate some sample data and fit a wide neural network to predict the target using a simple dense layer (1 x 512) -> Tanh activation -> linear layer (512 x 1)

```python
# @markdown #### Generate the sample dataset
SEED=2021
set_seed(seed=SEED)
n_samples = 32
inputs = torch.linspace(-1.0, 1.0, n_samples).reshape(n_samples, 1)
noise = torch.randn(n_samples, 1) / 4
targets = torch.sin(pi * inputs) + noise
plt.figure(figsize=(8, 5))
plt.scatter(inputs, targets, c='c')
plt.xlabel('x (inputs)')
plt.ylabel('y (targets)'); sns.despine()
plt.show()
```

```python
class RegNet(nn.Module):
    def __init__(self):
        ncells=512;
        super().__init__();
        self.layers = nn.Sequential(
        nn.Linear(1,ncells),
        nn.Tanh(),
        nn.Linear(ncells,1))
        
    def forward(self, x):
        return self.layers(x)

reg_net = RegNet()
loss_fun = nn.MSELoss();
lr = 0.003;
sgd_optim = torch.optim.SGD(reg_net.parameters(), lr=lr, momentum = 0.9); #what is momentum?



loss_list =[]
for epoch in range(1847):
    sgd_optim.zero_grad();
    y_pred = reg_net(inputs);
    loss = loss_fun(targets, y_pred);
    loss_list.append(loss.item());
    
    loss.backward()
    sgd_optim.step();
    
    
f, ax=plt.subplots(1,2,figsize=(15,3));
ax[0].plot(inputs, targets,'or',label='data');
ax[0].plot(inputs, y_pred.detach(),'--b',label='prediction')
ax[0].set_xlabel('Input'); ax[0].set_ylabel('Target');
ax[0].legend(frameon=False)

ax[1].plot(loss_list); sns.despine();
ax[1].set_xlabel('Epoch'); ax[1].set_ylabel('MSE loss');
ax[1].set_title('Its weird that the loss is not monotonic, on zooming in around 1-100 epoch');

plt.show();
```

#### Section 2: Learning hyperparameters:
In a NN, hyperparameters of interest are:
* **Learning rate**: 
    1) Low LR = Takes longer for convergence, but steady decrease usually
    2) High LR = Error might bounce around
* **Initialization**: In deep networks, things blow up or shrink very quickly
    1) Deep networks highly susceptible to initialization 
    2) Roughly want to maintain same variance across each layer
* **Depth of the network**: 
    1) Shallow networks converge slowly but surely
    2) Deep networks typically have a sigmoid like shape for convergence
