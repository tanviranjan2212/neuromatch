```python
import numpy as np
import matplotlib.pyplot as plt
import pdb
import math
import seaborn as sns
import torch.optim as optim
from torch import nn
from torch.utils.data import DataLoader, TensorDataset

import random
from IPython.display import display

import random
import torch
from tqdm import tqdm

def set_seed(seed=None, seed_torch=True):
    if seed is None:
        seed = np.random.choice(2 ** 32)
    random.seed(seed)
    np.random.seed(seed)
    if seed_torch:
        torch.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.cuda.manual_seed(seed)
        torch.backends.cudnn.benchmark = False
        torch.backends.cudnn.deterministic = True
    print(f'Random seed {seed} has been set.')


# In case that `DataLoader` is used
def seed_worker(worker_id):
    worker_seed = torch.initial_seed() % 2**32
    np.random.seed(worker_seed)
    random.seed(worker_seed)
SEED = 2021
set_seed(seed=SEED)
```

#### Section 1.2: Building MLPs in Pytorch

```python
class Net(nn.Module):
    def __init__(self, actv, input_feature_num, hidden_unit_nums, output_feature_num):
        super(Net,self).__init__()
        self.input_feature_num = input_feature_num;
        self.hidden_unit_nums = hidden_unit_nums;
        self.output_feature_num = output_feature_num;
        
        self.mlp = nn.Sequential();
        
        curr_in_num = input_feature_num;
        for i in range(len(hidden_unit_nums)):
            curr_out_num = hidden_unit_nums[i];
            layer = nn.Linear(curr_in_num, curr_out_num);
            curr_in_num = curr_out_num
            
            self.mlp.add_module(f'Linear_{i}', layer)
            
            actv_layer = eval('nn.%s'%actv)
            self.mlp.add_module(f'Activation', actv_layer)
            
        out_layer = nn.Linear(curr_in_num, output_feature_num);
        self.mlp.add_module('Output', out_layer)
    
    def forward(self, x):
        return self.mlp(x)
    
input = torch.zeros([100,2]);
net = Net('LeakyReLU(0.1)', input_feature_num=2, hidden_unit_nums=[100,10,5],
         output_feature_num=1)
y = net(input);
print(f'Output shape is {y.shape} for input shape {input.shape}')
```

##### Classification with MLPs using a final softmax layer and cross-entropy loss
The explainer video for this section is one of my favorite since it delves into a very simple derivation! Here's a summary in my words:

* The MLP networks we just built is great, but how can we use its output for classification. To classify data into C classes, we need the output layer to give us the probability of each training set to be in one of C classes. The **soft-max** layer helps us convert the output into such a distribution by doing the following transform:

$$ q_i = \frac{e^{Z_i}}{\sum_{j=0}^{C} e^{Z_j}} $$

And the cross-entropy loss between the correct label $Y_k$ for a training vector $k$ and the output distribution $\textbf{q}$ is 

\begin{equation}
L_k = \sum_{i=0}^C - Y_k log(q_{ki}) \\
\end{equation}

Since only one of $Y_k \neq 0$, that probability gets picked to be maximized
\begin{equation}
L_k = \sum_{i=0}^C - Y_k log(q_{ki}) \\
= -log (q_{k})
\end{equation}

Geometrically, this cross-entropy loss is just the sum of entropy of the true labels and the KL divergence of the true label and the softmax probability output

\begin{equation}
L_k = \sum_{i=0}^C - Y_k log(q_{ki}) \\
= \sum_{i=0}^C - Y_k log(Y_k) + \sum_{i=0}^C - Y_k log(\frac{Y_k}{q_{ki}}) \\
\end{equation}

KL-Divergence is symmetric, positive, but does not follow triangle inequality

##### Implement cross-entropy loss

```python
def cross_entropy_loss(x, labels):
    x_of_label = np.zeros_like(labels);
    
    for i, label in enumerate(labels):
        x_of_labels = x[i, label];
    
    losses = -x_of_labels + torch.log(torch.sum(torch.exp(x), axis=1))
    avg_loss = 1/len(labels)*torch.sum(losses)
    return avg_loss

labels = torch.tensor([0, 1])
x = torch.tensor([[10.0, 1.0, -1.0, -20.0],  # correctly classified
                  [10.0, 10.0, 2.0, -10.0]])  # Not correctly classified
CE = nn.CrossEntropyLoss()
pytorch_loss = CE(x, labels).item()

my_loss = cross_entropy_loss(x, labels)
print(f'Torch CE loss: {pytorch_loss}, my function loss: {my_loss}')
```

##### Create a spiral-looking dataset for classification

```python
def create_spiral_dataset(K,sigma, N):
    t=torch.linspace(0,1,N);
    X=torch.zeros(K*N,2); y=torch.zeros(K*N); #this way of declaration is important
    
    for k in range(K):
        X[k*N:(k+1)*N,0] = t*(torch.sin(2*math.pi/K*(2*t+k-1)) + torch.randn(N)*sigma);
        X[k*N:(k+1)*N,1] = t*(torch.cos(2*math.pi/K*(2*t+k-1)) + torch.randn(N)*sigma);
        y[k*N:(k+1)*N] = k
        
    return X,y

K = 4
sigma = 0.16
N = 1000

set_seed(seed=SEED)
X, y = create_spiral_dataset(K, sigma, N)
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
        
```

##### Create test-train splits, load dataset and do classification with MLP

```python
def create_test_train_splits(X,y,seed):
    torch.manual_seed(seed);
    N=X.shape[0]
    
    #First shuffle data, then choose random indices for test
    shuffled_indices = torch.randperm(N);
    X_shuffled, y_shuffled = X[shuffled_indices,], y[shuffled_indices]
    
    idx_test = torch.round(torch.tensor(N*0.2)).int();
    X_test, y_test, X_train, y_train = X_shuffled[:idx_test], y_shuffled[:idx_test], X_shuffled[idx_test:], y_shuffled[idx_test:]
    
    return X_test, y_test, X_train, y_train

X_test, y_test, X_train, y_train = create_test_train_splits(X, y, seed=SEED)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
plt.title('Test data'); sns.despine()
plt.show()
```

```python
g_seed = torch.Generator()
g_seed.manual_seed(SEED)

batch_size = 128
test_data = TensorDataset(X_test, y_test)
test_loader = DataLoader(test_data, batch_size=batch_size,
                         shuffle=False, num_workers=2,
                         worker_init_fn=seed_worker,
                         generator=g_seed)

train_data = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_data, batch_size=batch_size, drop_last=True,
                        shuffle=True, num_workers=2,
                         worker_init_fn=seed_worker,
                         generator=g_seed)
```

##### Train-test 

```python
def train_test_classification(net, criterion, optimizer, train_loader, test_loader,
                             num_epochs=1, verbose=False, training_plot=True,device='cpu'):
    
    net.train(); 
    train_losses = torch.zeros(num_epochs)
    
    for i_epoch in tqdm(range(num_epochs),position=0,leave=True):
        running_loss = [];
        for _, dat in enumerate(train_loader, 0):
            inputs, labels = dat;
            inputs = inputs.to(device).float();
            labels = labels.to(device).long();
            
            optimizer.zero_grad();
            output = net.forward(inputs);
#             pdb.set_trace();
            loss = criterion(output, labels);
            
            loss.backward();
            optimizer.step();
            
            train_losses[i_epoch] = train_losses[i_epoch]+loss.item();
            
    net.eval();
    def test(data_loader):
        correct, total = 0, 0
        for i, data in enumerate(data_loader):
            inputs, labels = data;
            inputs = inputs.to(device).float();
            labels = labels.to(device).long();
        
            outputs = net(inputs);
            
            _, predicted_labels = torch.max(outputs,1);
            
            total = total+labels.size(0);
            correct = correct+(predicted_labels==labels).sum().item();
            
        acc = 100*correct/total
        return acc, total
        
    train_acc, train_total = test(train_loader);
    test_acc, test_total = test(train_loader);
    
    if verbose:
        print(f"Accuracy on the {train_total} training samples: {train_acc:0.2f}")
        print(f"Accuracy on the {test_total} testing samples: {test_acc:0.2f}")

    if training_plot:
        plt.plot(train_losses)
        plt.xlabel('Batch')
        plt.ylabel('Training loss')
        plt.show()

    return train_acc, test_acc

set_seed(SEED);
net = Net('ReLU()',X_train.shape[1],[128],K).to('cpu'); #tip: X_train.shape[1] or X_train.size(0)

criterion = nn.CrossEntropyLoss();
optimizer = optim.Adam(net.parameters(), lr=1e-3);

num_epochs = 1000;
train_test_classification(net, criterion, optimizer, train_loader, test_loader,
                             num_epochs=num_epochs, verbose=False, training_plot=True,device='cpu')


```

For this we will work with the animal faces dataset

```python
# @title Download of the Animal Faces dataset
# @markdown Animal faces consists of 16,130 32x32 images belonging to 3 classes
import requests, os
from zipfile import ZipFile

print("Start downloading and unzipping `AnimalFaces` dataset...")
name = 'AnimalFaces32x32'
fname = f"{name}.zip"
url = f"https://osf.io/kgfvj/download"
r = requests.get(url, allow_redirects=True)
with open(fname, 'wb') as fh:
    fh.write(r.content)

with ZipFile(fname, 'r') as zfile:
    zfile.extractall(f"./{name}")

if os.path.exists(fname):
    os.remove(fname)
else:
    print(f"The file {fname} does not exist")

os.chdir(name)
print("Download completed.")
```

```python
def run_depth_optimizer(max_param_count, max_hidden_layer):
    
    def count_parameters(model):
#         pdb.set_trace()
        par_count = 0;
        for param in model.parameters():
            if param.requires_grad:
                par_count+=torch.numel(param) #important sn
        return par_count
    
    hidden_layers = np.arange(1,max_hidden_layer+1);
    
    test_scores = [];
    
    for hidden_layer in hidden_layers:
        hidden_units = np.ones(hidden_layer, dtype=np.int);
        
        wide_net = Net('ReLU()', X_train.shape[1], hidden_units,K).to('cpu');
        par_count = count_parameters(wide_net);
        
#         pdb.set_trace()
        while(par_count <= max_param_count):
            hidden_units = hidden_units +1;
            wide_net = Net('ReLU()', X_train.shape[1], hidden_units, K).to('cpu');
            par_count = count_parameters(wide_net); 
#             print(hidden_units)
#             print(par_count)
            
        criterion = nn.CrossEntropyLoss();
        optimizer = optim.Adam(wide_net.parameters(),lr=1e-3);
        
        _, test_acc = train_test_classification(wide_net, criterion, optimizer, train_loader, test_loader,
                             num_epochs=100, verbose=False, training_plot=False,device='cpu')
        test_scores += [test_acc]
        
    return hidden_layers, test_scores

set_seed(seed=SEED)
max_par_count = 100
max_hidden_layer = 5

hidden_layers, test_scores = run_depth_optimizer(max_par_count, max_hidden_layer)
plt.xlabel('# of hidden layers')
plt.ylabel('Test accuracy')
plt.plot(hidden_layers, test_scores)
plt.show()
```

Why the trade-off?

The model is not over-fitting necessarily since the number of parameters is the same. This decrease in accuracy is most likely caused due to the vanishing gradients issue
